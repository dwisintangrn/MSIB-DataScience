#Feature Engineering & EDA

```python
df_disease = df[(df['target'] == 1 )]
df_disease
```

```python
df_disease['sex'].value_counts()
```

```python
df_disease['sex'].value_counts().plot.pie(autopct='%.2f')
```

```python
bins2 = [0, 20, 50, 100]
labels2 = ['Remaja', 'Dewasa', 'Lansia']
```

```python
df['age_group'] = pd.cut(df['age'], bins=bins2, labels=labels2)
df.insert(1, 'age_group', df.pop('age_group'))
df
```

```python
df_disease["age_group"].value_counts(normalize = True).sort_values(ascending = True).plot(kind = "bar")
```

```python
sns.ecdfplot(data = df,
             x = "age",
             hue = "target")
```

```python
sns.distplot(df['age'], kde=True)
```

```python
sns.distplot(df_disease['age'], kde=True)
```

```python
mapping = {
    0: 'typical angina',
    1: 'atypical angina',
    2: 'non-anginal pain',
    3: 'asymptomatic'
}

df['chest_pain_type'] = df['chest_pain'].map(mapping)

df.insert(4, 'chest_pain_type', df.pop('chest_pain_type'))
df
```

```python
df_disease["chest_pain_type"].value_counts(normalize = True).sort_values(ascending = True).plot(kind = "barh")
```

```python
mapping = {
    0: 'upsloping',
    1: 'horizontal',
    2: 'downsloping'
}

df['st_type'] = df['st_slope'].map(mapping)

df.insert(12, 'st_type', df.pop('st_type'))
df
```

```python
df_st = df_disease.loc[:, ["st_depression", "st_slope", "st_type", "target"]]
df_st
```

```python
df_st["st_type"].value_counts(normalize = True).sort_values(ascending = True).plot(kind = "barh")
```

```python
sns.lmplot(data = df_st,
                x = "st_depression",
                y = "st_slope")
```

```python
mapping = {
    1: 'normal',
    2: 'fixed defect',
    3: 'reversable defect'
}

df['status_thalassemia'] = df['thalassemia'].map(mapping)

df.insert(14, 'status_thalassemia', df.pop('status_thalassemia'))
df
```

```python
df_disease["status_thalassemia"].value_counts(normalize = True).sort_values(ascending = True).plot(kind = "bar")
```

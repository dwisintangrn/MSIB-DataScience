#Menghitung BMI
- Membuat program untuk menghitung BMI sesuai rumus
- Mengupdate dataframe dengan menambahkan kolom hasil perhitungan BMI
- Mengkategorikan status BMI dengan kondisi tertentu
- Mengupdate dataframe dengan menambahkan kolom status BMI

```python
#Nama kolom diubah untuk memudahkan dalam memproses data
df_dataset.columns = ["Id", "Nama", "Jurusan", "Weight", "Height"]
df_dataset
```

```python
#Merubah tinggi badan dari cm ke m
df_dataset['Height_m'] = df_dataset['Height'] / 100
df_dataset
```

```python
#Menghitung BMI dan menempatkan hasilnya pada kolom baru
df_dataset['BMI'] = df_dataset['Weight'] / (df_dataset['Height_m'] ** 2)
df_dataset
```

```python
#Mendefinisikan status BMI
def status_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi <= 24.9:
        return 'Normal'
    elif 25 <= bmi <= 29.9:
        return 'Overweight'
    elif 30 <= bmi <= 34.9:
        return 'Obesity'

# Menambahkan kolom status BMI pada dataframe
df_dataset['status BMI'] = df_dataset['BMI'].apply(status_bmi)
df_dataset
```
